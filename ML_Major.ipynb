```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn import metrics
```

```python
scaler = StandardScaler()
scaler.fit(df)
scaled_ds = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
scaled_ds.head()
```

```python
pca = PCA(n_components=2)
pca.fit(scaled_ds)
PCA_ds = pd.DataFrame(pca.transform(scaled_ds), columns=(["col1","col2"]))
PCA_ds
```

```python
PCA_ds.describe().T
```

```python
from yellowbrick.cluster import KElbowVisualizer
kmean = KMeans(random_state=42)
visualizer = KElbowVisualizer(kmean, k=(2,10))
visualizer.fit(scaled_ds)
visualizer.show()
```

k = 4

```python
kmeans_model = KMeans(n_clusters = 4)
kmeans_model.fit_predict(scaled_ds)
pca_ds_kmeans = pd.concat([PCA_ds, pd.DataFrame({'cluster': kmeans_model.labels_})], axis=1)
```

```python
plt.figure(figsize=(8,6))
ax=sns.scatterplot(x="col1",y="col2",hue="cluster",data=pca_ds_kmeans, palette='dark')
plt.title("Clustering using K-Means Method")
plt.show()
```

```python
cluster_centers = pd.DataFrame(data=kmeans_model.cluster_centers_,columns=[df.columns])
```

```python
cluster_centers = scaler.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data=cluster_centers,columns=[df.columns])
cluster_centers
```

```python
df.reset_index(drop=True, inplace=True)
label_df = pd.concat([df,pd.DataFrame({'Cluster':kmeans_model.labels_})],axis=1)
label_df
```

```python
for c in label_df.drop(['Cluster'],axis=1):
    grid= sns.FacetGrid(label_df, col='Cluster', hue='Cluster',palette='dark')
    grid= grid.map(plt.hist, c)
plt.show()
```

Market segmentation is categorized into 4 group customers

Group 1 :

Customers in this group show a very low credit balance with a high balance frequency, indicating that they often use their credit up to the limit. They make significant one-time and installment purchases but rarely engage in cash advances. They tend to pay off their credit balance in full and have a short tenure, demonstrating an active but short-term credit usage pattern.

Group 2 :

Customers in this group have a low credit balance with a very high balance frequency, indicating that they frequently utilize their credit limit in full. They make very few purchases, both one-time and installment. Their cash advance transaction frequency is high, but the amount of cash advance and purchases is low. Their credit limit and payments are very low, with a long tenure, suggesting they tend to focus more on cash advances rather than purchases.

Group 3 :

Customers in this group have a very high credit balance with a very low balance frequency, indicating that they rarely utilize their credit balance in full. They are active in one-time purchases, installment purchases, and cash advances, with very high purchase transactions and a very high credit limit. They pay off their balance in full with very high values and have a very short tenure, indicating aggressive credit usage and strong repayment ability.

Group 4 :

Customers in this group have a very high credit balance with a low balance frequency, indicating that they rarely utilize their credit balance. They tend to engage in cash advances with very high frequency and value, but make very few one-time and installment purchases. They have high credit limits and payments but rarely pay off their balance in full, with a long tenure, suggesting a heavy credit usage pattern focused on cash advances with minimal purchases.
